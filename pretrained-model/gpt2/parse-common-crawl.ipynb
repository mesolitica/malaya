```python
from glob import glob
files = glob('feather/*.feather')
len(files)
```

```python
texts = []
```

```python
import pandas as pd
from tqdm import tqdm

for f in tqdm(files):
    df = pd.read_feather(f)
    texts.extend(df['sentence'])
```

```python
len(texts)
```

```python
texts[-1000:]
```

```python
with open('dumping-commmon-crawl.txt', 'w') as fopen:
    fopen.write('\n'.join(texts))
```
